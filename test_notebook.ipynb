```python
import matplotlib.pyplot as plt

# Bar Plot Örneği
categories = ['Python', 'JavaScript', 'Java', 'C++', 'Go']
values = [45, 38, 32, 28, 25]

df_plot = pd.DataFrame({'Dil': categories, 'Popülarite': values})

# Bar plot oluştur
plt.figure(figsize=(10, 6))
plt.bar(df_plot['Dil'], df_plot['Popülarite'], color='steelblue')
plt.xlabel('Programlama Dili')
plt.ylabel('Popülarite Skoru')
plt.title('Programlama Dillerinin Popülaritesi')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
